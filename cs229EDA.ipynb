```python
import numpy as np
import re

def clean_budget_value(budget_value):
    if pd.isna(budget_value):
        return np.nan

    if isinstance(budget_value, (int, float, np.integer, np.floating)):
        return float(budget_value)

    budget_str = str(budget_value).strip()
    if budget_str in {'', '-', '--'}:
        return np.nan

    clean_str = budget_str.replace(',', '')
    numbers = re.findall(r'\d+(?:\.\d+)?', clean_str)
    if not numbers:
        return np.nan
    if len(numbers) == 1:
        return float(numbers[0])

    return (float(numbers[0]) + float(numbers[1])) / 2

merged_df['budget_dinner'] = merged_df['budget_dinner'].apply(clean_budget_value).astype(float)
merged_df['budget_lunch'] = merged_df['budget_lunch'].apply(clean_budget_value).astype(float)
```

```python
merged_df.loc[:, 'store_id'] = (merged_df.loc[:, 'store_id'])
merged_df['budget_dinner'] = pd.to_numeric(merged_df['budget_dinner'], errors='coerce')
merged_df['budget_lunch'] = pd.to_numeric(merged_df['budget_lunch'], errors='coerce')
```

```python
def extract_genres(genre_str):
    if isinstance(genre_str, str):
        return [g.strip() for g in genre_str.split('、') if g.strip()]
    return []

merged_df['genres_list'] = merged_df['genre'].apply(extract_genres)
print("Added 'genres_list' column to merged_df:")
print(merged_df[['genre', 'genres_list']].head(10))
```

```python
all_unique_genres = set()
for genres in merged_df['genres_list']:
    for genre in genres:
        all_unique_genres.add(genre)

all_unique_genres_list = list(all_unique_genres)
print(f"Total unique genres found: {len(all_unique_genres_list)}")
print("First 10 unique genres:")
print(all_unique_genres_list[:10])
```

```python
# ---
# The genres seem to be pre-defined. As a result, the
# '・' character isn't as troll as it may seem. They are broken into like
# real categories, so I don't think it makes sense to clean.

from collections import Counter

genre_counts = Counter()

for genres in merged_df['genres_list']:
    genre_counts.update(genres)

genre_frequencies = pd.Series(genre_counts).sort_values(ascending=False)

print("Top 20 most frequent genres:")
print(genre_frequencies.head(50))
```

```python
N = 20
top_n_genres = genre_frequencies.head(N).index.tolist()

print(f"Selected top {N} genres for one-hot encoding:")
print(top_n_genres)
```

```python
for genre in top_n_genres:
    col_name = f'genre_{genre}'
    merged_df[col_name] = merged_df['genres_list'].apply(lambda x: 1 if genre in x else 0)

print("One-hot encoding for top N genres completed. Displaying head of relevant columns:")
print(merged_df[[f'genre_{g}' for g in top_n_genres[:5]]].head())
```

### Logistic Regression

```python
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score

target = "survived"

drop_cols = ["store_id", "name", "address", "genres_list", "genre"]
drop_cols = [c for c in drop_cols if c in merged_df.columns]

X = merged_df.drop(columns=[target] + drop_cols)
y = merged_df[target].astype(int)

numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()

preprocessor = ColumnTransformer([
    ("num", Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]), numeric_cols),
    ("cat", Pipeline([
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ]), categorical_cols)
])

clf = Pipeline([
    ("preprocess", preprocessor),
    ("model", LogisticRegression(penalty=None, max_iter=5000))
])


acc_scores = cross_val_score(clf, X, y, cv=5, scoring="accuracy")
f1_scores = cross_val_score(clf, X, y, cv=5, scoring="f1_macro")

print("CV accuracy scores:", acc_scores)
print("Mean CV accuracy:", acc_scores.mean())
print("CV test error (from accuracy):", 1 - acc_scores.mean())

print("CV f1_macro scores:", f1_scores)
print("Mean CV f1_macro:", f1_scores.mean())
print("CV test error (from f1_macro):", 1 - f1_scores.mean())
```

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
target = "survived"

drop_cols = ["store_id", "name", "address", "genres_list", "genre"]
drop_cols = [c for c in drop_cols if c in merged_df.columns]

X = merged_df.drop(columns=[target] + drop_cols)
y = merged_df[target].astype(int)

numeric_cols = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()

preprocessor = ColumnTransformer([
    ("num", Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]), numeric_cols),
    ("cat", Pipeline([
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ]), categorical_cols)
])

clf = Pipeline([
    ("preprocess", preprocessor),
    ("model", RandomForestClassifier())
])


acc_scores = cross_val_score(clf, X, y, cv=5, scoring="accuracy")
f1_scores = cross_val_score(clf, X, y, cv=5, scoring="f1_macro")

print("CV accuracy scores:", acc_scores)
print("Mean CV accuracy:", acc_scores.mean())
print("CV test error (from accuracy):", 1 - acc_scores.mean())

print("CV f1_macro scores:", f1_scores)
print("Mean CV f1_macro:", f1_scores.mean())
print("CV test error (from f1_macro):", 1 - f1_scores.mean())
```

```python
clf.fit(X, y)

feature_names = clf.named_steps["preprocess"].get_feature_names_out()
coefs = clf.named_steps["model"].coef_[0]

importance = (
    pd.DataFrame({"feature": feature_names, "coef": coefs})
    .assign(abs_coef=lambda d: d["coef"].abs())
    .sort_values("abs_coef", ascending=False)
)

print(importance.head(30))
```
